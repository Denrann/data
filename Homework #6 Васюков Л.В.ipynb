**Задание 2**. Анализ и удаление пропусков. Общее количество пропусков позволяет применить dropna, а не удаление столбца целиком.

```python
df.columns = [col.replace('No ','no').replace(' ','_').replace('__','_').lower() for col in df.columns]
df.info()
```

```python
df = df.dropna()
df.info()
```

**Задание 3**. Средний доход купивших Camry

```python
df1 = df[df['company'].isin(['Toyota']) & df['model'].isin(['Camry'])]
print('Средний доход купивших Camry равен', df1['annual_income'].mean())
```

**Задание 4**. Top-10 самых дорогих марок

```python
df2 = df.groupby(['company', 'model'])['price_($)']. max().reset_index().sort_values(by='price_($)', ascending=False).head(10)
df2[['company', 'model', 'price_($)']]
```

**Задание 5**. Средняя стоимость по типу кузова

```python
df3 = df.groupby('body_style')['price_($)']. mean()
df3
```

**Задание 6**. 10 клиентов с высоким доходом
10 клиентов с низким доходом

```python
df4 = df.sort_values(by='annual_income', ascending=False).head(10)
print('Список клиентом с высоким доходом:')
print(df4[['customer_name', 'annual_income', 'company', 'model']])

df5 = df.sort_values(['annual_income', 'customer_name'], ascending=[True, True]).head(10)
print('Список клиентом с низким доходом:')
print(df5[['customer_name', 'annual_income', 'company', 'model']])
```
