```python
def factorial(n):
    f = 1
    for x in range(1, n+1):
        f *= x
    return f
def P_rep(n,m):
    add = 1
    for x in m:
      add *= factorial(x)
    return int(factorial(n)/ add)
def A(n, m):
    return int(factorial(n) / factorial(n - m))
```

**Задание1.** Сколькими способами можно расставить на первой горизонтали шахматной доски комплект белых фигур (король, ферзь, два слона, два коня и двеладьи)?




```python
print(f'Существует {P_rep(8, [2,2,2,1,1])} способов')
```

```python
print(f'Существует {A(6,3)} способов')
```
