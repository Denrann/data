**Задание 1**
Имеется 2 выборки: первая выборка — это участки, обработанные агротехническим методом А;
вторая выборка — участки, обработанные агротехническим методом Б. Значения в выборках — это
урожайность. Требуется выяснить, является ли один из методов эффективнее другого, или
различия урожайности обусловлены случайными факторами.

Для проверки гипотезы применить t-критерий и перестановочный критерий.



```python
# H0 - различия урожайности обусловлены случайныи факторами
# H1 - один из методов эффективнее другого

import scipy.stats as sps
from tqdm.notebook import tqdm

A_1 = [3352, 3080, 3195, 3448, 3373, 2804, 3190, 2969, 2979, 3082, 3028, 3290, 3152, 3024, 3088,
3066, 3298, 2959, 3062, 2829, 2489, 3130, 3172]
B_1 = [3214, 3967, 3036, 3411, 3353, 3783, 3767, 3438, 3494, 3178, 2904, 3313, 3439, 3707, 3700,
3303, 3324, 3137, 3045, 2973, 3887, 3272, 3290, 3086, 3594, 2996, 3346, 3176, 3496, 3272]
alpha = 0.05

ttest_pvalue = sps.ttest_ind(A_1, B_1, equal_var=False, alternative='two-sided').pvalue

permut_pvalue = sps.ttest_ind(A_1, B_1, permutations=1000000, alternative='two-sided').pvalue

print(f'Один из методов эффективнее другого. Гипотеза H0 отвергается. p_value {ttest_pvalue:.5f} или {permut_pvalue:.5f} меньше {alpha}')
```

**Задание 2.**
В компании имеется три команды продаж. Необходимо определить, есть ли значительная разница
в их эффективности. В выборках представлены еженедельные объемы продаж за один и тот же
период времени.


```python
# 1. H0 - значительная разница в эффективности отсутсвует. H1 - имеется разница . Для применения F - критерия необходимо определить, имеется ли равенство дисперсий.
# 2. Для проверки равенства дисперсий применяем критерий Левене. H0.1 - равенство дисперсий, H1.1 - равенство не наблюдается.
import scipy.stats as sps

alpha = 0.05 # уровень значимости
data1 = [352.9, 312.0, 329.4, 367.2, 356.0, 270.7, 328.5, 295.5, 296.9, 312.3, 304.3, 343.6, 322.8,
303.7, 313.3, 310.0, 344.8, 293.8, 309.4, 274.4, 223.4, 319.6, 325.9, 277.7, 368.1]

data2 = [261.8, 321.8, 312.5, 381.3, 378.8, 326.2, 335.1, 284.5, 240.8, 306.1, 326.3, 369.2, 368.1,
304.5, 307.9, 278.1, 263.2, 251.7, 398.0, 299.6, 302.5, 269.9, 351.1, 255.4, 311.5]

data3 = [249.7, 307.4, 267.0, 236.9, 288.7, 309.3, 293.0, 303.6, 261.5, 273.7, 259.7, 273.8, 253.4,
212.3, 298.0, 271.9, 216.6, 310.8, 249.2, 292.3, 322.8, 295.8, 341.3, 234.4, 308.1]

stat, p = sps.levene(data1, data2, data3)
print(f'Гипотеза H0.1 о равенстве дисперсий не отвергается. p_value = {p:.2f} больше {alpha}. F-критерий можно применять.')
stat, p_f = sps.f_oneway(data1, data2, data3)
print(f'Гипотеза H0 отверагется, поскольку p-value {p_f:.4f} меньше {alpha}. Значиетльная разница в эфективности имеется.')

```
