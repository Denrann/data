**Задание 4**
Напишите программу-игру, в которой компьютер загадывает число от 1 до 100,а пользователь пытается его угадать. После каждой попытки программа должна сообщать, больше или меньше загаданное число, чем введенное пользователем.Игра продолжается до тех пор, пока число не будет угадано.

```python
import random
i = random.randint(1,100)
c = int(input('Введите число от 1 до 100: '))
while c != i:
  if c > i:
    print('Загаданное число меньше.')
  elif c < i:
    print('Загаданное число больше.')
  c = int(input('Введите число: '))
if c == i:
    print('Победа! Было загадано число',i, 'и Вы его отгадали.')
```

```python
# Для проверки на случайных числах убрать знак # и закомментировать массив [-2,1,-3,4,-1,2,1,-5,-4]
# from random import randint
# i = int(input())
# nums = [randint(-10,11) for _ in range(i)]
nums = [-2,1,-3,4,-1,2,1,-5,-4]
c = len(nums)+1
mx_sum = sum(nums)
mx_podm = nums
for n in range(c):
  for m in range(c):
    if m <= n:
        continue
    pr = sum(nums[n:m])
    if pr > mx_sum:
      mx_podm = nums[n:m]
      mx_sum = sum(mx_podm)
print('Непрерывным подмассивом массива', nums, 'c максимальной суммой',mx_sum, 'является', mx_podm)
```
